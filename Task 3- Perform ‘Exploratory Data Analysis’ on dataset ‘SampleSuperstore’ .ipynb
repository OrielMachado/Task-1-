```python
# Univariate analysis of numerical variables.
fig, ax = plt.subplots(2,2,figsize=(35,20))

plt.subplot(2,2,1)
sns.distplot(df["Profit"])
plt.xticks(fontsize=21)
plt.title("Profit",fontsize=40)

plt.subplot(2,2,2)
sns.distplot(df["Sales"])
plt.xticks(fontsize=21)
plt.title("Sales",fontsize=40)

plt.subplot(2,2,3)
sns.distplot(df["Discount"])
plt.xticks(fontsize=21)
plt.title("Discount",fontsize=40)

plt.subplot(2,2,4)
sns.distplot(df["Quantity"])
plt.xticks(fontsize=21)
plt.title("Quantity",fontsize=40)
```

```python
#looking at the Catagorical variables

Area=df[["Region","State","City","Quantity"]]
Area=Area.sort_values(by=["Region","State","City"],ascending=True, axis=0)
Area.head(15)
```

```python
#Plotting no. of Products sold per Region.

plt.style.use("seaborn")
Region=Area.groupby("Region").sum()
Region.plot(kind="barh",figsize=(10,5))
plt.ylabel("Region",fontsize=15)
plt.xlabel("Quantity of Products sold",fontsize=15)
```

West Region sells the most number items, followed by East and Central Regions.
The items sold by South Region amount the lowest quantity. 

```python
#Plotting no. of items sold per State.

plt.style.use("seaborn")
State=Area.groupby("State").sum()
State.plot(kind="barh",figsize=(50,25))
plt.yticks(fontsize=21) 
plt.xticks(fontsize=21) 
plt.ylabel("States",fontsize=40)
plt.xlabel("Quantity of Items Sold",fontsize=40)
  


```

The state of California sells the most amount of items, followed by the states of New York, Texas and Pennsylvania.
West Virginia ranks last on this list.

Los Angeles, San Francisco and San Diego contribute to most of the items sold in California.

#### Comparing the Sales, Discounts and Profit

```python
#Region

a=df.groupby("Region")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)
plt.style.use("seaborn")
a.head(15).plot(kind='bar')
```

The west has the highest profit as well as sales, followed by the east.
The Central region has the lowest profit even though its sales are higher than the South.
The Central region also gives the highest discount


```python
#State..Highest profit

a=df.groupby("State")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)

a.head(25).plot(kind='bar', figsize=(20,10))
plt.title('States making highest Profit', fontsize=25)

a.tail(24).plot(kind='bar', figsize=(20,10))
plt.title('States making highest loss', fontsize=25)
```

Califorina records the highest profit as well as sales, followed by New York and Washington.
California also the highest discount.

In total, 10 states run into losses.
Texas records the highest loss, followed by, Ohio and Pennsylvania. 

```python
#City
a=df.groupby("City")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)

a.head(40).plot(kind='bar', figsize=(15,5))
plt.title('Citys making highest profit', fontsize=25)

a.tail(40).plot(kind='bar', figsize=(15,5))
plt.title('Citys making highest loss', fontsize=25)
```

New York City has the highest profit as well as sales record, followed by, Los Angeles, Seattle and San Francisco.
Atlanta makes a high profit in comparison to its sales.
In comparison to a high sales record, Philadelphia incurred the greatest loss.It also records the highest discount given.
Houston, San Antonio and Lancaster  also incurred heavy losses.

```python
#Category

a=df.groupby("Category")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)
a.head(25).plot(kind='bar', figsize=(20,10))
```

Technology records the highest profit as well as sales.
Even though Office Supplies have lower sales as compared to Furniture, they make more profit.
Also, Technology gives the least discount whereas Office Supplies provide the highest.

```python
#Sub-Category

a=df.groupby("Sub-Category")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)
a.head(25).plot(kind='bar', figsize=(20,10))
```

Tables have made the most losses even though their sales are compariatively high, followed by, bookcases and supplies.
Copiers have shown to be most profitable ones, followed by phones and accessories.
Phones have the highest sales, followed by chairs and storage.
Binders give the highest discount.


```python
#ShipMode

a=df.groupby("Ship Mode")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)
a.head(25).plot(kind='bar', figsize=(20,10))
```

Standard class  records the highest sales and profit despite giving the highest discount.
Although the sales of Second Class are higher than First Class their profits are similar.
Products shipped on the same day make the least profit.

```python
#Segment

a=df.groupby("Segment")["Sales","Discount","Profit"].sum()
a=a.sort_values(by="Profit", ascending=False)
a.head(25).plot(kind='bar', figsize=(20,10))
```

Consumer Segment records the highest Profit as well as sales and discount, followed by, Corporate and Home Office.

Major conclusions:
    ..Discount does not make a huge impact on profit as well as sales.
    ..Even though California records the highest profit per state, New York City in New York holds the recorf for highest
      profit per city. 
    ..Shipping on the same day is the least profitable.
    ..Phones and chairs have high demand however, copiers, papers and envolope make most profit per product.

Proposed solutions to make more profit:

..Limit the discount given.
..Reduce sales in the citys of Philadelphia and Houston.
..Increase sales in New York City, Los Angeles and Seattle.
